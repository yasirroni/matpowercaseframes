```python
import numpy as np
from matpower import start_instance

from matpowercaseframes import CaseFrames
```

```python
def find_inf(df):
    num = df.select_dtypes(include="number")
    mask = np.isinf(num)  # DataFrame mask (same shape as num)
    r, c = np.where(mask.to_numpy())
    hits = list(zip(num.index[r], num.columns[c], num.to_numpy()[r, c]))
    return hits, mask
```

```python
m = start_instance()
```

```python
CASE_NAME = "case9241pegase.m"
cf_case9241pegase = CaseFrames(CASE_NAME)
```

```python
cf_case9241pegase.infer_numpy()  # optional, RuntimeWarning due to inf, expected
cf_case9241pegase.gen
```

```python
hits, mask = find_inf(cf_case9241pegase.gen)
cf_case9241pegase.gen.loc[mask.any(axis=1)]
```
